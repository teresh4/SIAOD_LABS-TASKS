–°–µ—Ä–µ–∂–∞ —Ö–æ—á–µ—Ç –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å —Å–≤–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º. –î–ª—è —ç—Ç–æ–≥–æ —É –Ω–µ–≥–æ –µ—Å—Ç—å —Å—Ç—Ä–æ–∫–∞ s‚Äâ=‚Äâs1s2... sn, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ n —Å–∏–º–≤–æ–ª–æ–≤. –ú–∞–ª—å—á–∏–∫ –ø—Ä–æ–≤–æ–¥–∏—Ç —Å–µ—Ä–∏—é –∏–∑ m —Ç–µ—Å—Ç–æ–≤. –í –∫–∞—á–µ—Å—Ç–≤–µ i-–≥–æ —Ç–µ—Å—Ç–∞ –æ–Ω –ø–æ–¥–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∞–ª–≥–æ—Ä–∏—Ç–º—É —Å—Ç—Ä–æ–∫—É slisli‚Äâ+‚Äâ1... sri (1‚Äâ‚â§‚Äâli‚Äâ‚â§‚Äâri‚Äâ‚â§‚Äân). –ö —Å–æ–∂–∞–ª–µ–Ω–∏—é, –µ–≥–æ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç —Å–ª–∏—à–∫–æ–º –¥–æ–ª–≥–æ, –ø–æ—ç—Ç–æ–º—É –°–µ—Ä–µ–∂–∞ –æ–±—Ä–∞—Ç–∏–ª—Å—è –∫ –≤–∞–º –∑–∞ –ø–æ–º–æ—â—å—é. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–µ—Å—Ç–∞ (li,‚Äâri) –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –æ—Ç—Ä–∞–±–æ—Ç–∞–µ—Ç –∞–ª–≥–æ—Ä–∏—Ç–º –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –Ω–∞ —ç—Ç–æ–º —Ç–µ—Å—Ç–µ –∏–ª–∏ –Ω–µ—Ç.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è –Ω–µ–ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ s, –µ–µ –¥–ª–∏–Ω–∞ (n) –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 105. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Å—Ç—Ä–æ–∫–∞ s —Å–æ—Å—Ç–æ–∏—Ç —Ç–æ–ª—å–∫–æ –∏–∑ —Å–∏–º–≤–æ–ª–æ–≤: 'x', 'y', 'z'.

–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ m (1‚Äâ‚â§‚Äâm‚Äâ‚â§‚Äâ105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–µ—Å—Ç–æ–≤. –°–ª–µ–¥—É—é—â–∏–µ m —Å—Ç—Ä–æ–∫ —Å–æ–¥–µ—Ä–∂–∞—Ç —Å–∞–º–∏ —Ç–µ—Å—Ç—ã. –í i-–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–∞ –ø–∞—Ä–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª li, ri (1‚Äâ‚â§‚Äâli‚Äâ‚â§‚Äâri‚Äâ‚â§‚Äân).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–î–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–µ—Å—Ç–∞ –≤—ã–≤–µ–¥–∏—Ç–µ ¬´YES¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫), –µ—Å–ª–∏ –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–∞–±–æ—Ç–∞–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–º —Ç–µ—Å—Ç–µ –∏ ¬´NO¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫) –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
zyxxxxxxyyz\
5\

1 3\

1 4\
3 6\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
YES\
YES\
NO\
YES\
NO\

```python
str = input()
m = int(input())
for i in range(m):
    mass = input().split()
    q = str(int(mass[0])-1, int(mass[1])-1)
    sub_s_len = 3
    if len(q) < sub_s_len:
        print("YES")
        continue
    i = 0
    while i < len(q):
        if i + sub_s_len <= len(q):
            s = q[i:i+sub_s_len]
            if "z" not in s or "y" not in s or "x" not in s:
                print("NO")
                break
        i += 1
    if i >= len(q):
        print('YES')
```

## –ó–∞–¥–∞—á–∞ 4
Chouti –∏ –µ–≥–æ –æ–¥–Ω–æ–∫–ª–∞—Å—Å–Ω–∏–∫–∞–º —Å–∫–æ—Ä–æ –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç –ø–æ—Å—Ç—É–ø–∞—Ç—å –≤ —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç. –ß—Ç–æ–±—ã –ø–æ–ø—Ä–æ—â–∞—Ç—å—Å—è –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º, –∫–ª–∞—Å—Å —É—Å—Ç—Ä–æ–∏–ª –±–æ–ª—å—à—É—é –ø—Ä–æ—â–∞–ª—å–Ω—É—é –≤–µ—á–µ—Ä–∏–Ω–∫—É, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–π –æ–¥–Ω–æ–∫–ª–∞—Å—Å–Ω–∏–∫–∏, —É—á–∏—Ç–µ–ª—è –∏ —Ä–æ–¥–∏—Ç–µ–ª–∏ –ø–µ–ª–∏ –∏ —Ç–∞–Ω—Ü–µ–≤–∞–ª–∏.

Chouti –≤—Å–ø–æ–º–Ω–∏–ª, —á—Ç–æ –≤ —ç—Ç–æ–π –≤–µ—á–µ—Ä–∏–Ω–∫–µ –ø—Ä–∏–Ω—è–ª–∏ —É—á–∞—Å—Ç–∏–µ ùëõ —á–µ–ª–æ–≤–µ–∫. –ß—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –≤–µ—á–µ—Ä–∏–Ω–∫—É —Å–º–µ—à–Ω–µ–µ, –∫–∞–∂–¥—ã–π –Ω–∞–¥–µ–ª –æ–¥–Ω—É —à–ª—è–ø—É –∏–∑ ùëõ —Ä–∞–∑–Ω–æ–≤–∏–¥–Ω–æ—Å—Ç–µ–π —Å—Ç—Ä–∞–Ω–Ω—ã—Ö —à–ª—è–ø —Å –Ω–æ–º–µ—Ä–∞–º–∏ 1,2,‚Ä¶ùëõ. –í–æ–∑–º–æ–∂–Ω–æ, —á—Ç–æ –æ–¥–Ω—É —Ä–∞–∑–Ω–æ–≤–∏–¥–Ω–æ—Å—Ç—å —à–ª—è–ø—ã –Ω–∞–¥–µ–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —á–µ–ª–æ–≤–µ–∫. –ù–µ–∫–æ—Ç–æ—Ä—ã–µ —Ä–∞–∑–Ω–æ–≤–∏–¥–Ω–æ—Å—Ç–∏ —à–ª—è–ø –º–æ–≥–ª–∏ –æ—Å—Ç–∞—Ç—å—Å—è –Ω–µ–≤–æ—Å—Ç—Ä–µ–±–æ–≤–∞–Ω–Ω—ã–º–∏ –∫–µ–º-–ª–∏–±–æ.

–ü–æ—Å–ª–µ –≤–µ—á–µ—Ä–∏–Ω–∫–∏ ùëñ-–π —á–µ–ª–æ–≤–µ–∫ —Å–∫–∞–∑–∞–ª, —á—Ç–æ –±—ã–ª–æ —Ä–æ–≤–Ω–æ ùëéùëñ –ª—é–¥–µ–π, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –±—ã–ª–∞ –Ω–∞–¥–µ—Ç–∞ —à–ª—è–ø–∞, –æ—Ç–ª–∏—á–∞—é—â–∞—è—Å—è –æ—Ç –µ–≥–æ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–π.

–ü—Ä–æ—à–ª–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ –¥–Ω–µ–π, –ø–æ—ç—Ç–æ–º—É Chouti –∑–∞–±—ã–ª, –∫—Ç–æ –±—ã–ª –≤ –∫–∞–∫–∏—Ö —à–ª—è–ø–∞—Ö, –Ω–æ –µ–º—É –æ—á–µ–Ω—å –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ —á—Ç–æ-—Ç–æ –ø—Ä–æ —ç—Ç–æ –≤—Å–ø–æ–º–Ω–∏—Ç—å! –ü—É—Å—Ç—å ùëèùëñ –±—É–¥–µ—Ç –≤–∏–¥–æ–º —Ç–æ–π —à–ª—è–ø—ã, –∫–æ—Ç–æ—Ä—É—é –Ω–∞–¥–µ–ª ùëñ-–π —á–µ–ª–æ–≤–µ–∫. Chouti —Ö–æ—á–µ—Ç, —á—Ç–æ–±—ã –≤—ã –Ω–∞—à–ª–∏ –∫–∞–∫–∏–µ-–Ω–∏–±—É–¥—å ùëè1,ùëè2,‚Ä¶,ùëèùëõ, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –ø—Ä–æ—Ç–∏–≤–æ—Ä–µ—á–∞—Ç —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏—è–º –∫–∞–∂–¥–æ–≥–æ —á–µ–ª–æ–≤–µ–∫–∞. –¢–∞–∫ –∫–∞–∫ —É –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –ª—é–¥–µ–π –º–æ–≥—É—Ç –±—ã—Ç—å –ø—Ä–æ–±–ª–µ–º—ã —Å –ø–∞–º—è—Ç—å—é, —Ä–µ—à–µ–Ω–∏—è –º–æ–∂–µ—Ç –Ω–µ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞—Ç—å.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ (1‚â§ùëõ‚â§105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª—é–¥–µ–π –Ω–∞ –≤–µ—á–µ—Ä–∏–Ω–∫–µ.

–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëé1,ùëé2,‚Ä¶,ùëéùëõ (0‚â§ùëéùëñ‚â§ùëõ‚àí1), –∏ –∑–∞–¥–∞—é—â–∞—è —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏—è –∫–∞–∂–¥–æ–≥–æ —á–µ–ª–æ–≤–µ–∫–∞.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–ï—Å–ª–∏ —Ä–µ—à–µ–Ω–∏—è –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, –≤—ã–≤–µ–¥–∏—Ç–µ ¬´Impossible¬ª.

–ò–Ω–∞—á–µ, –≤—ã–≤–µ–¥–∏—Ç–µ ¬´Possible¬ª –∞ –∑–∞—Ç–µ–º ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëè1,ùëè2,‚Ä¶,ùëèùëõ (1‚â§ùëèùëñ‚â§ùëõ).

–ï—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ä–µ—à–µ–Ω–∏–π, –≤—ã –º–æ–∂–µ—Ç–µ –≤—ã–≤–µ—Å—Ç–∏ –ª—é–±–æ–µ –∏–∑ –Ω–∏—Ö.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
3\
0 0 0\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
Possible\
1 1 1 \

–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
4\
0 1 2 3\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
Impossible

```python
n = int(input())
string = input().split()
a = {}
for i in range(n):
    num = n - int(string[i])
    if num not in a.keys():
        a[num] = 1
    else:
        a[num] += 1
hat = 1
ans = ''
count = 0
for key in a.keys():
    for i in range(key):
        ans += str(hat) + " "
        count += 1
    hat += 1
    if count > n:
        print("Impossible")
        return
print("Possible")
print(ans)
```

–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ n (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ100000) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–∞–Ω—Ü–∏–π.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω—ã n —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª p1, p2, ..., pn (1‚Äâ‚â§‚Äâpi‚Äâ‚â§‚Äân) ‚Äî —Ç–µ–∫—É—â–∞—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –º–µ—Ç—Ä–æ. –í—Å–µ —ç—Ç–∏ —á–∏—Å–ª–∞ —Ä–∞–∑–ª–∏—á–Ω—ã.

–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –≤–æ–∑–º–æ–∂–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã–≥–æ–¥–Ω–æ—Å—Ç–∏.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
3\
2 1 3\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
9

```python
n=int(input())
p=list(map(int,input().split()))
for i in range(n):
    p[i]-=1
color=[0]*n
c=1
s=[]
for i in range(n):
    if color[i]==0:
        j,k=i,1
        color[i]=c
        while p[j]!=i:
            j=p[j]
            color[j]=c
            k+=1
        c+=1
        s.append(k)
s.sort()
if len(s)==1:
    x=s[0]**2
else:
    x=(s[-1]+s[-2])**2
    for i in range(len(s)-2):
        x+=s[i]**2
print(x)
```

5\

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

1 3 1 2 1

```python
n = int(input())
l = list(map(int,input().split()))
r = list(map(int,input().split()))
a = []
for i in range(n):
    a.append(n - l[i] - r[i])
for i in range(n):
    for j in range(i):
        l[i] -= (a[j] > a[i])

for i in range(n):
    for j in range(i+1, n):
        r[i] -= (a[j] > a[i])
possible = True
for i in range(n):
    if l[i] != 0 or r[i] != 0:
        possible = False
        break
if possible:
    print("YES")
    print(a)
else:
    print("NO")
```

–ü–æ–¥—Å—Ç—Ä–æ–∫–æ–π s[i... j] (1‚Äâ‚â§‚Äâi‚Äâ‚â§‚Äâj‚Äâ‚â§‚Äâ|s|) —Å—Ç—Ä–æ–∫–∏ s = s1s2... s|s| –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞, —Ä–∞–≤–Ω–∞—è sisi‚Äâ+‚Äâ1... sj. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ s[2...4] —Å—Ç—Ä–æ–∫–∏ s = ¬´abacaba¬ª —Ä–∞–≤–Ω–∞ ¬´bac¬ª.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–∞ –Ω–µ–ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ s, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ —Å—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞ ('a' ... 'z'). –î–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏ s –Ω–µ –±–æ–ª–µ–µ 2000 —Å–∏–º–≤–æ–ª–æ–≤.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í—ã–≤–µ–¥–∏—Ç–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä –Ω–µ–ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –ø–æ–¥—Å—Ç—Ä–æ–∫-–ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–≤ –∑–∞–¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏.

–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–ø–µ—Ü–∏—Ñ–∏–∫–∞—Ç–æ—Ä %lld –¥–ª—è —á—Ç–µ–Ω–∏—è –∏–ª–∏ –∑–∞–ø–∏—Å–∏ 64-—Ö –±–∏—Ç–æ–≤—ã—Ö —á–∏—Å–µ–ª –Ω–∞ –°++. –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–æ—Ç–æ–∫–∏ cin, cout –∏–ª–∏ —Å–ø–µ—Ü–∏—Ñ–∏–∫–∞—Ç–æ—Ä %I64d.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
aa\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
1\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
aaa\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
5

```python
def pali(s):
    reverse = s[::-1]
    for i in range(len(s)):
        if s[i] != reverse[i]:
            return False
    return True
string = input()
n = len(string)
count = 0
for i in range(n-1):
    for j in range(i+1, n):
        s = string[i:j]
        if pali(s):
            for f in range(j, n):
                for k in range(f, n):
                    s1 = string[j:k+1]
                    if pali(s1) and len(s1) > 0:
                            count += 1
print(count)
```

–í —Å–ª–µ–¥—É—é—â–∏—Ö n —Å—Ç—Ä–æ–∫–∞—Ö –∑–∞–ø–∏—Å–∞–Ω—ã –ø–æ m —á–∏—Å–µ–ª ‚Äî —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ç–∞–±–ª–∏—Ü—ã. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —á–∏—Å–ª–∞ –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –æ–±—Ä–∞–∑—É—é—Ç –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ m.

```python
import copy
def check(p):
    bads = 0
    for i in range(len(p)):
        bads += p[i] != i
    return bads <= 2

n, m = map(int, input().split())
matrix = []
for i in range(n):
    matrix.append(list(map(int, input().split())))
    
for i in range(m):
    for j in range(i):
        flag = True
        for k in range(n):
            p = copy.copy(matrix[k])
            if not check(p):
                p[i], p[j] = p[j], p[i]
                if not check(p):
                    flag = False
        if flag:
            print("YES")
            return
print("No")
```
